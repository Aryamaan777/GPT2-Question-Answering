```
from aitextgen import aitextgen
from aitextgen.TokenDataset import TokenDataset
from aitextgen.tokenizers import train_tokenizer
from aitextgen.utils import GPT2ConfigCPU
#from aitextgen.colab import mount_gdrive, copy_file_from_gdrive
```

```
ai = aitextgen(tf_gpt2="124M") #to_gpu=True)
```

```
file_name="question_answer_pairs_modified.txt"
```

```
ai.train(file_name,
         line_by_line=False,
         from_cache=False,
         num_steps=5000,
         generate_every=1000,
         save_every=1000,
         save_gdrive=False,
         learning_rate=1e-4,
         batch_size=1, 
         )
```

```
val_file_name="10%.txt"
```

```
data=open(val_file_name,"r")
prev=""
q=""
a=""
questions=[]
answers=[]
#lines=[data.readline(),data.readline(),data.readline(),data.readline()]
for line in data:
  if(line.startswith("[QUESTION]:")):
    prev="ques"
    q+=line
  elif(line.startswith("[ANSWER]:")):
    prev="ans"
    a+=line
  elif(line.startswith("<|startoftext|>")):
    prev=""
    q=""
    a=""
    continue
  elif(line.startswith("<|endoftext|>")):
    prev=""
    questions.append(q)
    answers.append(a)
    continue
  elif(prev=="ans"):
    a+=line
  elif(prev=="ques"):
    q+=line
  print(line)
print(questions,answers)
```

```
import string
import re

def normalize_answer(s):
    """Lower text and remove punctuation, articles and extra whitespace."""
    def remove_articles(text):
        return re.sub(r'\b(a|an|the)\b', ' ', text)

    def white_space_fix(text):
        return ' '.join(text.split())

    def remove_punc(text):
        exclude = set(string.punctuation)
        return ''.join(ch for ch in text if ch not in exclude)

    def lower(text):
        return text.lower()

    return white_space_fix(remove_articles(remove_punc(lower(s))))
```

```
for i in range(len(answers)):
  answers[i]=normalize_answer(answers[i])
print(len(answers))
```

```
predictions=[]
for i in range(len(answers)):
  print(i/len(answers))
  strx=ai.generate(n=1,prompt=questions[i],max_length=35,return_as_list=True)
  #print(str(i)+":"+strx[0])
  l=len(questions[i])
  posa=strx[0].find("[ANSWER]")
  pos=strx[0].find("\n",posa)
  predictions.append(strx[0][l:pos])
print(len(predictions))
```

```
from collections import Counter

def f1_score(prediction, ground_truth):
    prediction_tokens = normalize_answer(prediction).split()
    ground_truth_tokens = normalize_answer(ground_truth).split()
    common = Counter(prediction_tokens) & Counter(ground_truth_tokens)
    num_same = sum(common.values())
    if num_same == 0:
        return 0
    precision = 1.0 * num_same / len(prediction_tokens)
    recall = 1.0 * num_same / len(ground_truth_tokens)
    f1 = (2 * precision * recall) / (precision + recall)
    return f1

def exact_match_score(prediction, ground_truth):
    return (normalize_answer(prediction) == normalize_answer(ground_truth))

"""def metric_max_over_ground_truths(metric_fn, prediction, ground_truth):
    score_for_ground_truth=0
    score = metric_fn(prediction, ground_truth)
    score_for_ground_truth=metric_fn(prediction,ground_truth)
    return max(score_for_ground_truth)"""


def evaluate(answers, predictions):
    f1 = exact_match = total = 0
    for i in range(len(predictions)):
      total += 1
      answer=answers[i]
      prediction = predictions[i]
      exact_match += exact_match_score(prediction, answer)
      f1 += f1_score(prediction, answer)

    exact_match = 100.0 * exact_match / total
    f1 = 100.0 * f1 / total

    return {'exact_match': exact_match, 'f1': f1}
```

```
res=evaluate(answers,predictions)
print(res)
```
